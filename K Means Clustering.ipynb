```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
# making artificial data using sklearn 
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)

# K-means clustering is an unsupervised machine learning algorithm. Hence the data only contains sample points 
# and no labels (i.e predicted output y).  
#n_samples : int, optional (default=100)
#    The total number of points equally divided among clusters.
    
#n_features : int, optional (default=2)
   #The number of features for each sample.

#centers : int or array of shape [n_centers, n_features], optional
    #(default=3)
    #The number of centers to generate, or the fixed center locations.

#cluster_std : float or sequence of floats, optional (default=1.0)
    #The standard deviation of the clusters.
```

```python
data
# data is a tuple which holds numpy arrays of the unlabeled points(data[0]) and the cluster index to which 
# they have been assigned(data[1])
# data[0] is a numpy matrix where each row respresents one unlabeled example and each column represents a feature
# Hence dimensions of data[0] is 200 X 2 
```

```python
np.shape(data[0])
```

```python
sns.set()
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
# plotting the data in 2-D .This can be done by plotting one feature against another.
# data[1] contains the mapping of each example to a color/cluster
# Here in this artificial example we do have labels for the inputs , but on a real dataset we usually won't.
# Here our aim would be to finally compare the clustering done by kmeans and original labels as shown below
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0]) # clustering the unlabeled inputs
```

```python
kmeans.cluster_centers_
# positions of the  4 cluster centroids after the algorithm has done its work
```

```python
kmeans.labels_ # the labels for each input example as predicted/computed by kmeans
# usually this is the last step we can do if we had been dealing with real world dataset
# since we have original labels we can compare
```

```python
data[1]
```

```python
fig , (ax1,ax2) = plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('KMeans')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap = 'rainbow',edgecolors='black')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow',edgecolors='black')
```
