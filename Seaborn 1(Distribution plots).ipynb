```python
import seaborn as sns
import matplotlib as plt
%matplotlib inline
plt.rcParams["patch.force_edgecolor"] = True
```

```python
tips = sns.load_dataset('tips') # seaborn comes with built-in datasets
```

```python
tips.head(10)
```

```python
# sns.distplot(tips['total_bill'],kde=False,hist_kws=dict(edgecolor="k", linewidth=1))

# distplot is used to plot the distribution of a univariate set of observations
sns.set()
sns.distplot(tips['total_bill'],kde=False,bins=30,hist_kws=dict(edgecolor="k", linewidth=1))
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=40) # the plot shows that majority of bill amounts are in the range of 10-20$
```

```python
sns.distplot(tips['size'],kde = False) # size of the party coming at restaurant is majorly 2
```

```python
sns.distplot(tips['tip'],kde = False) # tip largely lies in the range of 2-4$,i.e majority people tip in this range only
```

```python
# iterating a dataframe
l = []
for index,row in tips.iterrows():
    if row['sex'] is 'Male':
        l.append(1) # l[index] = 1--> gives an error for the very first insertion since the list is still empty or so
    else:
        l.append(0)
print(l)  
```

```python
sns.distplot(l,kde = False) # to plot the sex distibution , first coverted the data into integer format
# majority population is male 
```

```python
# jointplot
# a jointplot basically gives us a plot comparing 2 properties/features in a dataset
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)  
# x, y are column names of the 2 props to be compared and data holds the df name
```

```python
# deafault kind is scatter plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind = 'reg') # gives a linear regression line
# its 2 distplot and a scatter plot in between showing the variance of data between 2 columns
# as bill amount increases , so does the tip 
```

```python
# pairplot
# pairwise plot/comparision of all numerical features in a dataframe
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue = 'sex')
```

```python
sns.pairplot(tips,hue='smoker')
```
