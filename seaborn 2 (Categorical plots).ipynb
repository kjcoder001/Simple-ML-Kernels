```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips) # x is a categorical data , y is numerical,data holds the dataframe
# shows that male tend to spend/eat more in the restaurant
# by deafault the estimator is the mean of both the columns . We can change this estimator according to our choice
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std) # estimator is standard devn 
```

```python
sns.countplot(x='sex',data=tips) # this is similar to barplot except the fact that y axis is already assigned to count
# which explicitly returns the count of the occurences in dataset of the category mentioned in x!
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
sns.barplot(x='smoker',y='tip',data=tips) # smokers ain't bad people after all! they tip equally!
```

```python
sns.barplot(x='smoker',y='total_bill',data=tips) # smokers eat more . Nicotine ain't no good
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips) # more informative than a boxplot. Its includes all the points that belongs 
# that category unlinke boxplot which leaves out outliers
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips) # combination of violinplot and stripplot 
# shows all the data points
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='bar') # genral type of plot, can mention any type in the kind key-value pair
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='point')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='strip')
```
